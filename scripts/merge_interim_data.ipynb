# Merge data all sites

```python
import pandas as pd
import os

def merge_to_interim():
    # 1. Thiết lập đường dẫn (dành cho Jupyter Notebook)
    # Vì Notebook của bạn nằm trong thư mục con 'interim preprocessing',
    # ta lấy thư mục cha của thư mục hiện tại để ra thư mục gốc (Root)
    current_dir = os.getcwd()
    base_dir = os.path.dirname(current_dir) 
    
    interim_dir = os.path.join(base_dir, 'data', 'interim')
    
    # Danh sách các file cần gộp
    file_names = ['batdongsan_interim.csv', 'thuviennhadat_interim.csv']
    file_paths = [os.path.join(interim_dir, name) for name in file_names]

    print(f"Thư mục gốc: {base_dir}")
    print("--- Bắt đầu quy trình gộp dữ liệu thành viên ---")

    dataframes = []
    for path in file_paths:
        if os.path.exists(path):
            df = pd.read_csv(path)
            df['source_site'] = os.path.basename(path).replace('_interim.csv', '')
            dataframes.append(df)
            print(f"Đã nạp: {os.path.basename(path)} | Số lượng: {len(df)} dòng")
        else:
            print(f"Cảnh báo: Không tìm thấy file {os.path.basename(path)} tại {interim_dir}")

    if not dataframes:
        print("Lỗi: Không tìm thấy bất kỳ dữ liệu nào để gộp!")
        return

    # 2. Gộp dữ liệu
    merged_df = pd.concat(dataframes, ignore_index=True)
    initial_count = len(merged_df)
    
    # 3. Loại bỏ trùng lặp
    duplicate_cols = ['tieu_de', 'dia_chi', 'gia']
    merged_df.drop_duplicates(subset=duplicate_cols, keep='first', inplace=True)
    
    print(f"\nTổng số dòng sau gộp: {initial_count}")
    print(f"Số lượng sau khi loại trùng: {len(merged_df)}")

    # 4. Lưu file tổng hợp vào lại thư mục interim
    output_filename = 'all_sites_interim.csv'
    output_path = os.path.join(interim_dir, output_filename)
    
    merged_df.to_csv(output_path, index=False, encoding='utf-8-sig')
    print(f"\n--- Hoàn tất! File đã lưu tại: {output_path} ---")
```

```python
merge_to_interim()
```
